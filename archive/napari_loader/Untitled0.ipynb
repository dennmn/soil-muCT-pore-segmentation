```python
from google.colab import drive
import os

checkpoints_dir = os.path.join(project_path, "checkpoints")

if not os.path.exists(checkpoints_dir):
    os.makedirs(checkpoints_dir)
    print(f"Created folder: {checkpoints_dir}")

urls = {
    "vit_h": "https://dl.fbaipublicfiles.com",
    "vit_l": "https://dl.fbaipublicfiles.com",
    "vit_b": "https://dl.fbaipublicfiles.com"
}

# הגדרת User-Agent שמדמה דפדפן Firefox עדכני
USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/115.0"

# הורדה באמצעות שורת פקודה (wget) עם דגל User-Agent
for model_name, url in urls.items():
    destination = os.path.join(checkpoints_dir, f"sam_{model_name}.pth")
    # בדיקה פשוטה אם הקובץ קיים ובגודל סביר (מעל 1MB)
    if os.path.exists(destination) and os.path.getsize(destination) > 1000000:
        print(f"File {os.path.basename(destination)} already exists and seems complete. Skipping download.")
    else:
        print(f"Downloading {os.path.basename(destination)} using wget with browser User-Agent...")
        # שימוש ב-! לפני הפקודה מריץ פקודות מערכת הפעלה (Bash) ב-Colab
        # הוספת הדגל -U (או --user-agent) עם המחרוזת של הדפדפן
        !wget -O "$destination" -U "$USER_AGENT" "$url"
        print(f"Finished processing {model_name}.")

print("All download processes attempted.")
```
